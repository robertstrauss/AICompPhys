```python
%matplotlib notebook

import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.color_palette('bright')

import torch
```

```python
def solve_ode(f, u0, t):
    u = np.zeros(((int(np.ceil((t[1]-t[0])/t[2])+1),)))
    u[0] = u0
    i = 0
    for tt in np.arange(t[0], t[1], t[2]):
        u[i+1] = u[i]+t[2]*f(u[i], tt)
        i+=1
    return u
```

```python
def f(x,t): return -1*x
x0 = 5#*np.ones((12,12))
t = (0, 10, 0.1)
```

```python
sol = solve_ode(f, x0, t)
```

```python
plt.figure(0)
plt.plot(sol[:])
```
